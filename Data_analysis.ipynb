### Scraping data at Mudah.my

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

motorList = []
priceList = []
conditionList = []
locationList = []
sellerList = []

i = 1

url = "https://www.mudah.my/Malaysia/Motorcycles-1040/r25-for-sale?o="+str(i)+"&q=r25&so=1&th=1"

while i<=15:
    
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'lxml')
    
    motor = soup.find_all('h2', {'class':'list_title'})
    price = soup.find_all('div', {'class':'ads_price'})
    condition = soup.find_all('font', {'class':'icon_label'})
    location = soup.find_all('div', {'class':'area'})
    seller = soup.find_all('div', {'class':'container_ico'})
    
    motor = [k.text for k in motor]
    motor = motor[:-1]
    
    price = [k.text for k in price]
    price = price[:-1]
    
    condition = [k.text for k in condition]
    condition = condition[:-1]
    
    location = [k.text for k in location]
    location = location[:-1]
    
    seller = [k.text for k in seller]
    seller = seller[:-1]
    
    motorList += motor
    priceList += price
    conditionList += condition
    locationList += location
    sellerList += seller
    
    print("number of pages that been scraped : ", i)
    i = i+1  
```

```python
df_motor = pd.DataFrame()
df_price = pd.DataFrame()
df_condition = pd.DataFrame()
df_location = pd.DataFrame()
df_seller = pd.DataFrame()
df_graphicCard = pd.DataFrame()

print(len(motorList))
print(len(priceList))
print(len(conditionList))
print(len(locationList))
print(len(sellerList))

df_motor['item_name'] = motorList
df_price['item_price'] = priceList
df_condition['item_condition'] = conditionList
df_location['item_location'] = locationList
df_seller['item_seller'] = sellerList
```

```python
df_motorList = pd.concat([df_motor, df_price, df_condition, df_location, df_seller], axis=1)
df_motorList
```

### Checking null values

```python
df_motorList.isnull().sum()
```

### Removing null values

```python
df_motorList.dropna(inplace= True)
df_motorList.shape
```

```python
df_motorList
```

```python
cleaned_df = df_motorList[~df_motorList['item_name'].str.contains(' naza ', case=False)]
cleaned_df.shape
```

### Change column datatype from String to Numeric

```python
cleaned_df['item_price'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
cleaned_df['item_price'] = pd.to_numeric(cleaned_df['item_price'])

cleaned_df
```

### Distinguish data based on year 

```python
motor_2019 = cleaned_df[cleaned_df['item_name'].str.contains('2019', case=False)]
motor_2018 = cleaned_df[cleaned_df['item_name'].str.contains('2018', case=False)]
motor_2017 = cleaned_df[cleaned_df['item_name'].str.contains('2017', case=False)]
motor_2016 = cleaned_df[cleaned_df['item_name'].str.contains('2016', case=False)]
motor_2015 = cleaned_df[cleaned_df['item_name'].str.contains('2015', case=False)]
motor_2014 = cleaned_df[cleaned_df['item_name'].str.contains('2014', case=False)]
motor_2013 = cleaned_df[cleaned_df['item_name'].str.contains('2013', case=False)]
```

## Visualization 

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = ('2019', '2018','2017','2016','2015', '2014', '2013')
x_axis = np.arange(len(x))
y_axis = [len(motor_2019), len(motor_2018), len(motor_2017), len(motor_2016), len(motor_2015), len(motor_2014), len(motor_2013)]
 
    
plt.figure(dpi=150, figsize=(12,6))
plt.bar(x_axis, y_axis, align='center', alpha=0.5, color='#9400D3')
plt.xticks(x_axis, x)
plt.ylabel('Count\n')
plt.xlabel('\nyear model')
plt.title('\n\nNumber of Yamaha R25 post in Mudah.my\n ')
 
plt.show()
```

```python
fig1, (ax1, ax2) = plt.subplots(1,2 ,  figsize=(14, 6))
#pie chart

condition_group = cleaned_df.groupby(['item_condition'])['item_name'].count().reset_index()

totalNew = condition_group.loc[0, 'item_name']
totalUsed = condition_group.loc[1, 'item_name']



labels = 'New', 'Used'
sizes = [totalNew, totalUsed]
explode = (0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

#-------------------------------------------------------------------------------------------------

#bar chart

seller_group = cleaned_df.groupby(['item_seller'])['item_name'].count().reset_index()

totalPrivate = seller_group.loc[0, 'item_name']
totalCompany = seller_group.loc[1, 'item_name']

x = ('Private Seller', 'Company')
x_axis = np.arange(len(x))
y_axis = [totalPrivate, totalCompany]
  
ax2.bar(x_axis, y_axis, align='center', alpha=0.5, color='#20B2AA')
ax2.set_xticks(x_axis)
ax2.set_ylabel('Count\n')
ax2.set_xlabel('\nType')
ax2.set_xticklabels(x)
ax2.set_title('\n\nNumber of sellers\n ')


display(condition_group, seller_group)

plt.show()
```
